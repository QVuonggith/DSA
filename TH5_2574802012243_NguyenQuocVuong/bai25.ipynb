Phát biểu bất biến tham lam
Khi một đỉnh u được lấy ra khỏi hàng đợi ưu tiên, khoảng cách ước lượng hiện tại của nó đã đạt giá trị tối ưu: d[u] = δ(s, u).

 Chứng minh (Phản chứng)
- Giả sử u là đỉnh đầu tiên bị chọn sai => d[u] > δ(s, u).
- Xét đường đi ngắn nhất thực tế từ nguồn s đến u. Gọi x là đỉnh cuối cùng đã tối ưu, và y là đỉnh kế tiếp chưa tối ưu trên đường đi đó (y có thể trùng u).
- Vì x đã tối ưu và cạnh (x, y) đã được cập nhật (relax), ta có: d[y] = δ(s, y).
- Vì đồ thị không có cạnh âm, đoạn đường từ y đến u có trọng số >= 0: δ(s, y) <= δ(s, u).
- Bắc cầu lại, ta được: d[y] = δ(s, y) <= δ(s, u) < d[u] => d[y] < d[u].
- Mâu thuẫn: Nếu d[y] < d[u], hàng đợi ưu tiên phải chọn y trước u. Do đó giả thiết sai => d[u] bắt buộc phải tối ưu.

Vì sao cần trọng số không âm?
- Ở bước chứng minh trên, ta đã dùng điều kiện trọng số >= 0 để khẳng định khoảng cách đến đỉnh dọc đường nhỏ hơn hoặc bằng đỉnh kết thúc: δ(s, y) <= δ(s, u).
- Nếu có cạnh âm, khoảng cách đi tiếp có thể giảm xuống (δ(s, y) > δ(s, u)). Lúc này d[y] có thể lớn hơn d[u], thuật toán sẽ lấy u ra trước và chốt kết quả sai, bỏ lỡ đường đi ngắn hơn qua y.

```python
import heapq

def dijkstra_demo(graph, start):
    # Khởi tạo khoảng cách ban đầu là vô cùng
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    
    # Hàng đợi ưu tiên (Min-Priority Queue) lưu (khoảng_cách, đỉnh)
    pq = [(0, start)]
    
    print(f"--- Bắt đầu Dijkstra từ đỉnh nguồn {start} ---")
    
    while pq:
        # Lấy đỉnh có khoảng cách nhỏ nhất ra khỏi hàng đợi
        current_distance, current_node = heapq.heappop(pq)
        
        # BẤT BIẾN THAM LAM: Ngay khi lấy ra, khoảng cách này được coi là tối ưu
        print(f"\n-> Lấy đỉnh '{current_node}' ra khỏi hàng đợi. Ước lượng hiện tại d[{current_node}] = {current_distance}")
        print(f"   [Bất biến] Chốt luôn khoảng cách ngắn nhất đến '{current_node}' là: {current_distance}")
        
        # Duyệt các đỉnh kề để cập nhật (Relaxation)
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            
            # Nếu tìm thấy đường đi ngắn hơn
            if distance < distances[neighbor]:
                old_dist = distances[neighbor]
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))
                print(f"   Cập nhật cạnh ({current_node} -> {neighbor}, trọng số {weight}): d[{neighbor}] giảm từ {old_dist} xuống {distance}")
                
    return distances

# ==========================================
# THỬ NGHIỆM 1: Đồ thị trọng số KHÔNG ÂM (Chạy ĐÚNG)
# ==========================================
graph_positive = {
    'A': {'B': 4, 'C': 1},
    'B': {'D': 1},
    'C': {'B': 2, 'D': 5},
    'D': {}
}
print("=== TRƯỜNG HỢP 1: ĐỒ THỊ TRỌNG SỐ DƯƠNG ===")
dijkstra_demo(graph_positive, 'A')


# ==========================================
# THỬ NGHIỆM 2: Đồ thị CÓ CẠNH ÂM (Chạy SAI)
# ==========================================
graph_negative = {
    'A': {'B': 3, 'C': 4},
    'B': {'D': 2},
    'C': {'B': -2, 'D': 5}, # Cạnh C -> B là cạnh âm
    'D': {}
}
print("\n\n=== TRƯỜNG HỢP 2: ĐỒ THỊ CÓ CẠNH ÂM ===")
dijkstra_demo(graph_negative, 'A')
```
