PHÁT BIỂU BẤT BIẾN VÒNG LẶP P(i):
Sau vòng lặp thứ i (với i chạy từ 0 đến n-2), đoạn mảng a[0..i] chứa i+1 phần tử nhỏ nhất của mảng và các phần tử này đã được sắp xếp theo đúng thứ tự tăng dần.

CHỨNG MINH TÍNH ĐÚNG ĐẮN:

a. Tính khởi tạo (Initialization):

Xét sau vòng lặp đầu tiên khi i = 0.

Thuật toán tìm phần tử nhỏ nhất trong toàn bộ mảng và đổi chỗ nó về vị trí a[0].

Lúc này, đoạn mảng a[0..0] chỉ có 1 phần tử duy nhất. Một phần tử thì hiển nhiên luôn đúng thứ tự và nó chính là phần tử nhỏ nhất của mảng.
-> Bất biến P(0) ĐÚNG.

b. Tính duy trì (Maintenance):

Giả sử bất biến đúng sau vòng lặp thứ i-1. Nghĩa là đoạn mảng a[0..i-1] đã chứa i phần tử nhỏ nhất của mảng và đã được sắp xếp tăng dần.

Ở vòng lặp thứ i, thuật toán sẽ tìm phần tử nhỏ nhất trong đoạn còn lại từ a[i] đến a[n-1].

Vì đoạn trước đó (a[0..i-1]) đã lấy hết i phần tử nhỏ nhất rồi, nên phần tử nhỏ nhất trong đoạn còn lại chắc chắn phải là phần tử nhỏ thứ i+1 của toàn mảng.

Thuật toán tiến hành đổi chỗ phần tử này về vị trí a[i]. Vì nó lớn hơn hoặc bằng tất cả các phần tử đã sắp xếp trước đó trong đoạn a[0..i-1], nên sau khi đổi chỗ, đoạn mảng a[0..i] sẽ chứa đúng i+1 phần tử nhỏ nhất và vẫn giữ nguyên thứ tự tăng dần.
-> Bất biến P(i) ĐÚNG (Tính duy trì được đảm bảo).

c. Tính hoàn thành (Termination):

Vòng lặp ngoài sẽ dừng lại sau khi kết thúc vòng lặp thứ i = n-2.

Áp dụng phát biểu bất biến tại bước này: Đoạn mảng a[0..n-2] sẽ chứa n-1 phần tử nhỏ nhất của mảng và đã được sắp xếp theo thứ tự tăng dần.

Khi n-1 phần tử đầu tiên đã là các phần tử nhỏ nhất và đã vào đúng vị trí tăng dần, thì phần tử duy nhất còn lại ở vị trí cuối cùng a[n-1] bắt buộc phải là phần tử lớn nhất mảng.

Do đó, toàn bộ mảng từ a[0] đến a[n-1] đã được sắp xếp hoàn chỉnh.

KẾT LUẬN: Thuật toán hoàn toàn đúng đắn và luôn dừng sau n-1 bước.

```python
def selection_sort(a):
    n = len(a)

    for i in range(0, n - 1):
        min_idx = i  
        for j in range(i + 1, n):
            if a[j] < a[min_idx]:
                min_idx = j 
                
        a[i], a[min_idx] = a[min_idx], a[i]
        
    return a
mang_chua_sap_xep = [64, 25, 12, 22, 11]
mang_da_sap_xep = selection_sort(mang_chua_sap_xep)

print("Mảng sau khi sắp xếp:", mang_da_sap_xep)
```
