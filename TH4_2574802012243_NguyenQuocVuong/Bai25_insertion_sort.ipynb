Phát biểu bất biến vòng lặp:
Bất biến P: "Trước mỗi vòng lặp ngoài thứ i, đoạn mảng a[0..i-1] chứa đúng các phần tử ban đầu của chính đoạn đó nhưng đã được sắp xếp theo thứ tự tăng dần."
Chứng minh:
Khởi tạo (i = 1): Đoạn a[0..0] chỉ có 1 phần tử nên luôn đúng.
Duy trì: Vòng lặp thứ i chèn a[i] vào đúng vị trí trong đoạn a[0..i-1]. Sau khi chèn, đoạn a[0..i] được sắp xếp xong. Khi i tăng lên 1 ở lượt kế tiếp, bất biến tiếp tục đúng.
Kết thúc (i = n): Vòng lặp dừng khi i = n. Thay i = n vào bất biến, ta có đoạn a[0..n-1] (toàn bộ mảng) đã được sắp xếp đúng.

```python
def insertion_sort(a):
    n = len(a)
    for i in range(1, n):
        # Bất biến: Đoạn a[0..i-1] luôn luôn đã được sắp xếp
        assert all(a[k] <= a[k+1] for k in range(i-1)), f"Sai tại i={i}"
        
        # Tiến hành chèn a[i]
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key
        
    # Kiểm tra lần cuối khi i = n (toàn bộ mảng đã xếp xong)
    assert all(a[k] <= a[k+1] for k in range(n-1)), "Mảng chưa xếp xong"
    return a

# Chạy thử
print(insertion_sort([5, 2, 4, 6, 1, 3]))
```
